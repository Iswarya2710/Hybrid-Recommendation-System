```python
plt.figure(figsize=(10, 6))
sns.countplot(x=df['Category'])
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Product Count by Category')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot( x=df['Price'], bins=10)
plt.xlabel('Price')
plt.ylabel('Count')
plt.title('Product Price Distribution')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x=df['Price'],y=df['Rating'])
plt.xlabel('Price')
plt.ylabel('Rating')
plt.title('Price vs Rating')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot( x=df['Category'], y=df['Price'])
plt.xlabel('Category')
plt.ylabel('Price')
plt.title('Product Price by Category')
plt.xticks(rotation=45)
plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# Plot 1: Bar plot of categories
axs[0, 0].set_title('Category Distribution')
sns.countplot(x=df['Category'], ax=axs[0, 0])
axs[0, 0].tick_params(axis='x', rotation=45)

# Plot 2: Histogram of prices
axs[0, 1].set_title('Price Distribution')
sns.histplot( x=df['Price'], kde=True, ax=axs[0, 1])

# Plot 3: Box plot of ratings
axs[1, 0].set_title('Rating Distribution')
sns.boxplot( x=df['Rating'], ax=axs[1, 0])

# Plot 4: Bar plot of colors
axs[1, 1].set_title('Color Distribution')
sns.countplot(x=df['Color'], ax=axs[1, 1])
axs[1, 1].tick_params(axis='x', rotation=45)

# Adjust the layout
plt.tight_layout()
#Display the plots
plt.show()
```

```python
columns = ['Price', 'Rating']

# Compute the correlation matrix
corr_matrix = df[columns].corr()

# Create a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')

# Customize the plot
plt.title('Correlation Heatmap')
plt.xticks(rotation=45)
plt.yticks(rotation=0)

# Display the plot
plt.show()
```
